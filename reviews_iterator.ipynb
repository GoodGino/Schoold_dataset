```python
#3.2 Term frequency
#TF = (Number of times term T appears in the particular row) / (number of terms in that row)
```

```python
tf1 = (data['review_body'][1:2]).apply(lambda x: pd.value_counts(x.split(" "))).sum(axis = 0).reset_index()
tf1.columns = ['words','tf']
tf1
```

```python
#3.3 Inverse Document Frequency
#IDF = log(N/n), where, N is the total number of rows and n is the number of rows in which the word was present
#The more the value of IDF, the more unique is the word.
```

```python
for i,word in enumerate(tf1['words']):
  tf1.loc[i, 'idf'] = np.log(data.shape[0]/(len(data[data['review_body'].str.contains(word)])))
tf1
```

```python
#3.4 Term Frequency – Inverse Document Frequency (TF-IDF)
#TF-IDF is the multiplication of the TF and IDF which we calculated above.
```

```python
tf1['tfidf'] = tf1['tf'] * tf1['idf']
tf1
```

```python
#preprocessing steps

#stemmer = PorterStemmer()
lemmatizer = nltk.WordNetLemmatizer()
stop = stopwords.words('english')
#To replace punctuation with a space use the following:
translation = str.maketrans(string.punctuation,' '*len(string.punctuation))


def preprocessing(line):
    tokens=[]
    line = line.translate(translation)
    line = nltk.word_tokenize(line.lower())
    for t in line:
        #if(t not in stop):
            #stemmed = stemmer.stem(t)
        stemmed = lemmatizer.lemmatize(t)
        tokens.append(stemmed)
    return ' '.join(tokens)
```

```python
def clean_review(review):
    result = ''
    for subreview in review:
#         letters_only = re.sub('[^a-zA-Z]', ' ', subreview)
#         words = letters_only.lower().split()
        words = subreview.lower().split()
        stopwords_eng = set(stopwords.words("english"))
        useful_words = [x for x in words if not x in stopwords_eng]
        useful_words_string = ' '.join(useful_words)
        result = result + useful_words_string
    print('ok')
    return(result)
```

```python
stemmer = PorterStemmer()
```

```python
def stem_words(words_list, stemmer):
    return [stemmer.stem(word) for word in words_list]
```

```python
def tokenize(text):
    tokens = word_tokenize(text)
    stems = stem_words(tokens, stemmer)
    return stems
```

```python
reviews_data = reviews_data.query('star_rating != 3')
reviews_data['clean_review'] = reviews_data['review_body'].apply(clean_review)
reviews_data['star_sentiment'] = reviews_data['star_rating'].apply(lambda x:1 if x > 3 else 0)



```

# build the model

```python
from sklearn.linear_model import SGDClassifier
from sklearn.feature_extraction.text import HashingVectorizer
from sklearn.feature_extraction import DictVectorizer
from sklearn.feature_selection import SelectKBest,chi2
```

```python
vectorizer = HashingVectorizer(decode_error='ignore', n_features = 2**18,
                               tokenizer = tokenize, non_negative=True)
```

```python
y_classes = np.array([0, 1])
max_iterations = 299
```

```python
clf_logit_sgd = SGDClassifier(loss = 'log', n_jobs = 2, learning_rate = 'optimal',
                              random_state = 12, verbose = 0, shuffle = True)
```

```python
validation_accuracy_list = []
```

```python
x_validation = vectorizer.transform(reviews_data['clean_review'])
y_validation = reviews_data['star_sentiment']
```

```python
for i, mini_batch in enumerate(reviews_iterator): 
    mini_reviews_data = mini_batch.copy().query('star_rating != 3')
    mini_reviews_data['clean_review'] = mini_reviews_data['review_body'].apply(clean_review)
    mini_reviews_data['star_sentiment'] = mini_reviews_data['star_rating'].apply(lambda x: 1 if x > 3 else 0)

    mini_reviews_data['star_sentiment']=mini_reviews_data['star_sentiment'].notnull()
    
    while '' in mini_reviews_data['clean_review']:
        mini_reviews_data['clean_review'].remove('')
    
    x_train = vectorizer.transform(mini_reviews_data['clean_review'])
   
    y_train = mini_reviews_data['star_sentiment']
    
    clf_logit_sgd.partial_fit(x_train, y_train, classes = y_classes)
    
    if i % 1 == 0:
        validation_accuracy = clf_logit_sgd.score(x_validation, y_validation)
        validation_accuracy_list.append((i, validation_accuracy))
        print ("Validation Accuracy at {0}:{1}".format(i, validation_accuracy))
        
    if i >= max_iterations:
        break
    
   
```
